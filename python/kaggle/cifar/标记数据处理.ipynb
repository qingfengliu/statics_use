```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
data=pd.read_csv(r'D:\书籍资料整理\kaggle\cifar-10\trainLabels.csv')
data.head()
```

```python
label=LabelEncoder()

label.fit(data['label'])
data['y']=label.transform(data['label'])
```

```python
data.head()
```

```python
data.to_csv(r'D:\书籍资料整理\kaggle\cifar-10\label.csv',index=False)
joblib.dump(label,r'D:\书籍资料整理\kaggle\cifar-10\label.pkl')
```
