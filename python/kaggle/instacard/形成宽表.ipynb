```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
import joblib
from sklearn.model_selection import train_test_split

aisles=pd.read_csv(r'D:\书籍资料整理\kaggle\instacard\aisles.csv')
departments=pd.read_csv(r'D:\书籍资料整理\kaggle\instacard\departments.csv')
order_products_prior=pd.read_csv(r'D:\书籍资料整理\kaggle\instacard\order_products__prior.csv')
order_products_train=pd.read_csv(r'D:\书籍资料整理\kaggle\instacard\order_products__train.csv')
orders=pd.read_csv(r'D:\书籍资料整理\kaggle\instacard\orders.csv')
products=pd.read_csv(r'D:\书籍资料整理\kaggle\instacard\products.csv')
```

```python
#将数据合成为大宽表
temp=pd.concat([order_products_prior,order_products_train])
print(temp.shape)
print(orders.shape)
data_all=orders.merge(temp,how='left',on='order_id')
data_all=data_all.merge(products,how='left',on='product_id')
data_all=data_all.merge(aisles,how='left',on='aisle_id')
data_all=data_all.merge(departments,how='left',on='department_id')
data_all.head()
```

```python
data_all.to_csv(r'D:\书籍资料整理\kaggle\instacard\宽表.csv',index=False)
```
